```
#colab to pandas
df= pd.read_csv("pima-indians-diabetes.csv")
df.columns = np.arange(9) #rename labels
df
```

```
X = df.iloc[:,0:8] #independent variable
Y = df.iloc[:,8] #dependent variable
```

```
funct = []
funct.append('tanh')
funct.append('sigmoid')
funct.append('relu')
```

```
analysisdf = pd.DataFrame(columns= ["first layer activation","second  layer activation", "third layer activation",
                                    "accuracy"])

analysisdf  
```

```
from numpy.random import seed 
from tensorflow import set_random_seed
from sklearn.preprocessing import *
rownumber = 0
for fun in funct:
    for func in funct:
        for functi in funct:
            set_random_seed(2)
            seed(7) 
            scaler = StandardScaler()
            X = scaler.fit_transform(X)
            model = Sequential()
            model.add(Dense(1050, input_dim=8, kernel_initializer='RandomNormal', activation=fun )) #bias_initializer='zeros'
            model.add(Dense(525, kernel_initializer='RandomNormal', activation=func))
            model.add(Dense(125, kernel_initializer='RandomNormal', activation=functi))
            model.add(Dense(1, kernel_initializer='RandomNormal', activation='sigmoid'))
            model.compile(loss='binary_crossentropy', optimizer='RMSprop', metrics=['accuracy'])
            history = model.fit(X, Y, epochs=150, batch_size=97,validation_split=0.20 ) #shuffle = False, 
            scores = model.evaluate(X, Y)
            analysisdf.loc[rownumber,"first layer activation"]=  fun
            analysisdf.loc[rownumber,"second  layer activation"]= func
            analysisdf.loc[rownumber,"third layer activation"]= functi
            analysisdf.loc[rownumber,"accuracy"]= (" \n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
            wks.set_dataframe(analysisdf,(1,1))
            rownumber += 1
            
```

To conclude for first, second ,third layer we can use activators tanh relu tanh for this model to get highest accuracy.
